```python
import numpy as np
import pandas as pd
from scipy.interpolate import griddata
from scipy.stats import uniform
import matplotlib.pyplot as plt

d = pd.read_csv("Howell1.csv",sep=';')
d2 = d[d.age >= 18]

def dunif(x, minimum=0,maximum=1,log=False):
    from scipy.stats import uniform
    if log==False:
        result=uniform.pdf(x=x,loc=minimum,scale=maximum-minimum)
    elif log==True:
        result=uniform.logpdf(x=x,loc=minimum,scale=maximum-minimum)
    return result

def dnorm(x,mean=0,sd =1,log=False):
    from scipy.stats import norm
    if log==False:
        result=norm.pdf(x,loc=mean,scale=sd)
    elif log==True:
        result=norm.logpdf(x,loc=mean,scale=sd)
    return result

n = 100
mu_list = np.linspace(150,160,n)
sigma_list = np.linspace(7,9,n)
post = np.array(np.meshgrid(mu_list,sigma_list)).reshape(2,-1).T
nrow = len(post)

post_LL = [sum(dnorm(d2.height,post[:,0][i],post[:,1][i],log=True)) for i in range(nrow)]

post_prod = [(post_LL[i] +
              dnorm(post[:,0][i],178,20,log=True) +
              dunif(post[:,1][i],0,50,log=True)) 
              for i in range(nrow)]

post_prob = np.exp(post_prod - max(post_prod))

X = np.linspace(post[:,0].min(), post[:,0].max(), n)
Y = np.linspace(post[:,1].min(), post[:,1].max(), n)
Z = griddata((post[:,0], post[:,1]), post_prob, (X[None,:], Y[:,None]))
plt.contour(X,Y,Z,color='red')
```

```python
import numpy as np
import matplotlib.pyplot as plt

post_p = post_prob/post_prob.sum()

sample_rows = np.random.choice(nrow,size=int(1e4),replace=True,p=(post_p))
sample_mu = post[:,0][sample_rows]
sample_sigma = post[:,1][sample_rows]

plt.plot(sample_mu, sample_sigma, '.', alpha=0.05)
plt.axis('equal')
plt.grid(False)
plt.xlabel('sample_mu', fontsize=10)
plt.ylabel('sample_sigma', fontsize=10)
```
